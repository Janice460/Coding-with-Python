```python
from itertools import accumulate
import operator
```

```python
a= [1,2,3,4]
accum = accumulate(a)
print(list(a))
print(list(accum))
```

```python
# func = max, operator.mul ...
accum2 = accumulate(a, func=operator.mul )
print(list(a))
print(list(accum2))
```

```python
from itertools import groupby
```

```python
a= [1,2,3,4]

def smaller_than_3(x):
    return x<3
group_obj = groupby(a, key=smaller_than_3)

for key, value in group_obj:
    print(key, list(value))
```

```python
persons = [
    {'name':'Tim', 'age': 35},
    {'name': 'Dan', 'age': 25},
    {'name':'Lisa', 'age': 25},
    {'name': 'Claire', 'age': 20}
]

group_obj2 = groupby(persons, key = lambda x: x['age'])
for key, value in group_obj2:
    print(key, list(value))
```

## Lambda
lambda expressions: expresion

```python
add10 = lambda x:x+10
add10(25)
```

```python
mult = lambda x,y: x*y
mult(20,23)
```

```python
points2D = [(1,2), (15,1), (5,-1), (10,4)]

def sort_by_y(x):
    return x[1]

points2D_sorted = sorted(points2D, key = sort_by_y)
points2D_sorted_1 = sorted(points2D, key = lambda x:x[1])
points2D_sorted_2 = sorted(points2D, key = lambda x:x[0]+x[1])

print(points2D)
print(points2D_sorted)
print(points2D_sorted_1)
print(points2D_sorted_2)
```

```python
# map(func, seq)
a = [1,2,3,4,5]
b = map(lambda x: x+2,a)
print(list(b))
```

```python
a = [1,2,3,4,5,6]
b = filter(lambda x: x%2==0, a)
print(list(b))

c = [x for x in a if x%2==0]
print(c)
```

```python
# reduce(func, seq), reduce a single value
from functools import reduce
a = [1,2,3,4,5,6]

product_a = reduce(lambda x, y: x+y,a)
print(product_a)
```

```python
class ValueTooHighError(Exception):
    pass

class ValueTooHSmallError(Exception):
    def __init__(self, message, value):
        self.message = message
        self.value = value

def test_value(x):
    if x > 100:
        raise ValueTooHighError('value es too high')
    if x<5:
        raise ValueTooHSmallError('value is too small', x)

try:
    test_value(3)
except ValueTooHighError as e:
    print(e)
except ValueTooHSmallError as e:
    print(e.message, e.value)
```

## JSON

```python
import json

person = {'name':'Johm',
          'age': 30,
          'city': 'New York',
          'hasChildren': False,
          'titles': ['engineer','programmer']}

personJSON = json.dumps(person, indent =4, sort_keys = True)
print(personJSON)

with open('person.json', 'w') as file:
    json.dump(person, file, indent =4)
```

```python
person = json.loads(personJSON)
print(person)
```

```python
with open('person.json', 'r') as file:
    person = json.load(file)
    print(person)
```
